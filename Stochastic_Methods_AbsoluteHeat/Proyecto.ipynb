<center><h4>II. Introduccion</h4></center>

<div style="text-align: justify">La fisica estadistica, tambien conocida como mecanica estadistica, propone tratar a los sistemas de manera estocastica. Contrario al enfoque determinista de la mecanica clasica, id est el comportamiento del sistema esta totalmente especificado; al tratar a los sistemas de manera estocastica no es posible conocer con certeza cual sera el siguiente estado que se tomara, dado que el sistema se rige apartir de los axiomas de la probabilidad, y por ende, solo es posible conocer la probabilidad de que este tome cierto estado. Formalmente, un sistema se describe de la manera siguiente</div><br>

<div style="text-align: justify"><i> Sea $X$ un sistema termodinamico con espacio de estados $S$, Temperatura $T$, Numero de particulas $N$ y volumen $V$.
La probabilidad de que las $N$ particulas de $X$ tomen el estado $Y \in \Omega \subset S$ es </i></div>

$$ \mathbb{P}[X[Y] \in \Omega]=\frac{\int_{\Omega}e^{-\beta H(X)}}{\int_{S} e^{-\beta H(X)}}$$

<i>donde $\beta=\frac{1}{\kappa_B T}$ es la beta termodinamica, $\kappa_B$ es la la constante de Boltzman y $H(X)$ es el hamiltoniano del sistema</i> [1]

<div style="text-align: justify"> Como se puede observar a apartir de la definicion, el comportamiento sistema depende del ente conocido como <i>Hamiltoniano</i>. El hamiltoniano, estrictamente hablando, es una funcion que depende del sistema termodinamico, es decir depende de las particulas (Posicion), la temperatura y volumen; la forma que el hamiltoniano adopte es lo que definira los distintos estados que el sistema tendra</div>

<center><h4>III. Distintos modelos de Hamiltoniano </h4></center>

<h5>3.1) Modelo de Ising</h5> <br>

<div style="text-align: justify">El modelo de ising es un modelo matematico para describir el comportamiento de un sistema termodinamico, este se basa en el <i>espin</i> de cada particula del sistema. En este modelo se propone que el espin puede tomar dos estados, -1 y 1, y partir de ello podemos definir el hamiltoniano que describira el sistema. Formalmente el modelo de Ising es el siguiente: </div><br>
 
<div style="text-align: justify"><i> Sea $x_i$ una particula del sistema termodinamico $X$, con posicion $P$, entonces el espin de la particula esta definido como  </i></div>
$$\sigma_i(P)=1, \; o \; \sigma_i(P)=-1$$
<div style="text-align: justify"><i>Ademas el hamiltoniano del sistema esta dado por </i></div>
$$H(X)=-\sum_{(i,j)}J\sigma_i \sigma_j - \sum_{j}h \sigma_j $$
<div style="text-align: justify"><i>donde $J$ representa la interaccion entre espines y $h$ la fuerza externa [2] </i></div><br>
    
Fisicamente, el espin de cada particula tiene un interpretacion. Por ejemplo, si se tratara de un material magnetico, el espin de cada particula dictaria la carga de la particula.

    
<h5>3.1) Modelo de Fermi-Dirac</h5>
    
El modelo de Fermi-Dirac en vez de centrarse en el espin del sistema, tiene su enfoque en la energia individual de cada estado del sistema. Formalmente tenemos lo siguiente.
    
<div style="text-align: justify"><i> Sea $s_i$ un estado del sistema termodinamico $X$, con energia $\varepsilon_i$, entonces el hamiltoniano de Fermi-Dirac esta dado por</i></div>
    
$$H(X)=\sum_{i}n_i \varepsilon_i$$

<div style="text-align: justify"><i> donde $n_i$ es el numero de particulas en el estado $s_i$ [3]</i></div>

<center><h4>IV. Nuestro modelo </h4></center>

<div style="text-align: justify">El modelo que usaremos sera una modificacion del modelo de <i>Maxwell-Boltzman</i> [4], donde cada estado $s_i$ sera la posicion que adopta una particula $x_j$. Definamos el modelo: </div><br>

<div style="text-align: justify"><i>Sea $X$ un sistema termodinamico aislado en forma de cuadrado de longitud $L$ , con $N$ particulas y temperatura $T$. Sea $S=[0,L] \times [0,L]$ el espacio de estados del sistema.<br> Definimos la energia del estado $s_i$ como</i></div>
    
$$\varepsilon(s_i)=E_c(s_i)+E_p(s_i)=E_c(s_i)=\frac{p_{i_x}^2+p_{i_y}^2}{2m} $$
    
<div style="text-align: justify"><i>Donde $m$ es la masa de la particula, $p_{i_x}, p_{i_y}$ son los momentos en la coordenada x Y y respectivamente<br>
Definimos el hamiltoniano del sistema como</i></div>
    
$$H(X)=\frac{1}{2m}\sum_{i=1}^N p_{i_x}^2+p_{i_y}^2$$

<div style="text-align: justify"><i>Por lo tanto, la probabilidad de encontrar al sistema en el estado $\Omega \subset S$ es</i></div>
    
$$ F=\frac{1}{Z} \int_\Omega e^{-\beta H(X)}$$
    
<div style="text-align: justify"><i> Por ultimo, definimos el calor especifico $C$ como </i></div>
    
$$C=\frac{1}{N^2 T}Var(H(X_i)) $$
    
<div style="text-align: justify"><i>Donde $X_i \sim \mathcal{L}(e^{-\beta H(X)})$ </i></div>

<center><h4>V. Implementacion </h4></center>

<div style="text-align: justify">Para implementar el modelo usaremos el algoritmo de <i>Metropolis-Hasting</i> [5] el cual es un algoritmo para simular de un kernel de probabilidad dado, apartir de la generacion de una cadena de Markov.<br>
El algoritmo es el siguiente </div><br>


<div style="text-align: justify"><i> Dada un kernel de probabilidad $f$, $P$ un kernel de transicion y $X_t \sim \mathcal{L}(f)$. Simulamos de $f$ de la siguiente manera <br>
1) Simular $Y_t \sim \mathcal{L}(P)$ y $U \sim Uniforme(0,1)$<br>
2) Si $U \leq min\left\{1,\frac{f(Y_t)P(Y_t,X_t)}{f(X_t)P(X_t,Y_t)}\right\}$, entonces $X_{t+1}=Y_t \sim \mathcal{L}(f)$ <br>
3) En caso contrario $X_{t+1}=X_t$</i></div> <br>

<div style="text-align: justify">Podria resultar evidente que para simular del sistema termodinamico nuestra eleccion de kernel de probabilidad sera la probabilidad de encontrar al sistema en un cierto estado, no obstante esto genera ciertas dificultades tecnicas al ser implementado (como el escoger de manera correcta el kernel de transicion, y mas aun: el hamiltoniano del sistema depende del estado); para resolver estas dificultades se formulara un proceso equivalente al cambio de estado de nuestro sistema.</div><br>

<div style="text-align: justify"><i> Dada una particula $e_t$ en la posicion $(X_t,Y_t)$, entonces el cambio en configuracion de estado de la particula $e_t$ en un cuadrado de lados $2 \alpha$ es equivalente a </i></div>

$$ X_{t+1}=X_t+\alpha \xi_1$$
$$Y_{t+1}=Y_t+\alpha \xi_2 $$

<div style="text-align: justify"><i>donde $\xi_i$ es un numero en $(-1,1)$</i></div><br>

Realizando lo anterior para cada particula, entonces es posible simular del sistema de la siguiente forma:

<div style="text-align: justify"><i>
1) Posicionamos las $N$ particulas en un estado $s_t$<br>
2) Simulamos $\xi_i \sim Uniforme(-1,1)$ y $U \sim Uniforme(0,1)$<br> 
3) Cambiamos el estado de cada particula al estado $l_{t}$ segun el esquema anterior<br>
4) Hacemos $s_{t+1}=l_t$ si $U \leq min \left\{1, e^{-\beta \Delta H(X)}\right\}$ <br>
5) En caso contrario $s_{t+1}=s_t$ </i></div> <br>

Por lo tanto, usando integracion Monte-Carlo, se tiene que

$$ C = \frac{L^2}{N^2 T M}\sum_{t=1}^M (s_t-E)^2 + O\left(\frac{1}{\sqrt{M}}\right)$$

donde $E= \mathbb{E}(s_t) \approx \frac{1}{M}\sum_{t=1}^M s_t$

El primer paso sera implementar el sistema de $N$ particulas.

```python
class Particula(object):
    
    def __init__(self,x,y,m):
        """
        Funcion inicializacion de particula
        x =: posicion x
        y =: posicion y
        m =: masa
        """
        
        self.x=x
        self.y=y
        self.m=m
    
    def positionX(self):
        """
        Funcion que regresa la posicion x
        """
        
        return self.x
    
    def positionY(self):
        """
        Funcion que regresa la posicion y
        """
        
        return self.y
    
    def changeP(self,x,y):
        """
        Funcion que cambia la posicion
        x =: posicion x
        y =: posicion y
        """
        
        self.x=x
        self.y=y
        
    def velocidad(self,x,y):
        """
        Funcion que regresa la velocidad
        """
        
        return np.linalg.norm([x-self.x,y-self.y],2)
    
    def momentum(self,x,y):
        """
        Funcion que regresa el momento lineal
        """
        
        return self.m*self.velocidad(x,y)
    
    def energia(self,x,y):
        """
        Funcion que regresa la energia
        """
        
        return (1/2)*(self.m*self.velocidad(x,y)**2)
```

```python
class Sistema(object):
    
    def __init__(self,L,T,N):
        """
        Funcion de inicializacion del Sistema
        L =: Longitud del sistema
        T =: Temperatura del sistema
        N =: Numero de particulas
        """
        
        self.L=L
        self.T=T
        self.N=N
        # Auxiliares
        self.Particulas=[]
        self.pastPositionX=[]
        self.pastPositionY=[]
        self.Initial=True
        
        for i in range(0,N):
            self.Particulas.append(Particula(0,0,1))
            self.pastPositionX.append(0)
            self.pastPositionY.append(0)
    
    def initSistema(self):
        """
        Funcion que inicializa la posicion cristalina inicial
        """
        
        aux1=0
        aux2=0
        count=0
            
        for i in range(0,self.N):
        
            if(count >= np.sqrt(self.N)):
                count=0
                aux1=0
                aux2=self.Particulas[i-1].positionY()+self.L/np.sqrt(self.N)
            elif(count!=0):
                aux1=self.Particulas[i-1].positionX()+self.L/np.sqrt(self.N)
            
            self.Particulas[i].changeP(aux1,aux2)
            self.pastPositionX.append(aux1)
            self.pastPositionY.append(aux2)
            
            count+=1
            
    def positions(self):
        """
        Funcion que nos regresa la posicion de cada particula
        """
        
        posX=[]
        posY=[]
        for i in range(0,self.N):
            posX.append(self.Particulas[i].positionX())
            posY.append(self.Particulas[i].positionY())
            
        return posX,posY
    
    def perturbateParticula(self,i,a,xi=[0,0]):
        """
        Funcion para perturbar el estado de la particula
        
        i =: Particula a perturbar
        a =: Libertad de movimiento
        xi =: Perturbacion
        """
        
        # Auxiliares
        self.pastPositionX[i]=self.Particulas[i].positionX()
        self.pastPositionY[i]=self.Particulas[i].positionY()
        
        # Condiciones de frontera
        if(self.Particulas[i].positionX()+a*xi[0] > self.L or self.Particulas[i].positionY()+a*xi[1] > self.L):
            return
        elif(self.Particulas[i].positionX()+a*xi[0] < 0 or self.Particulas[i].positionY()+a*xi[1] < 0):
            return
        
        # Perturbamos segun el esquema
        self.Particulas[i].changeP(self.Particulas[i].positionX()+a*xi[0],self.Particulas[i].positionY()+a*xi[1])
        
    def perturbateParticles(self,a,xi):
        """
        Funcion para perturbar el sistema
        
        a =: Libertad de movimiento
        xi =: Perturbacion del sistema
        
        """
        for i in range(0,self.N):
            # Perturbamos la particula i del sistema
            self.perturbateParticula(i,a,xi[i])
    
    def returnPast(self):
        """
        Funcion para regresar al estado anterior
        """
        
        for i in range(0,self.N):
            self.Particulas[i].changeP(self.pastPositionX[i],self.pastPositionY[i])
        return;
    
    def Energia(self):
        """
        Funcion que calcula la energia total del sistema
        """
        
        # Si el sistema esta nuevo
        if(self.Initial):
            self.Initial=False
            return 0
        
        aux=0
        for i in range(0,self.N):
            # Calculamos la energia de la particula i
            aux+=self.Particulas[i].energia(self.pastPositionX[i],self.pastPositionY[i])
        return aux
    
    def returnN(self):
        """
        Funcion que regresa el numero de particulas
        """
        
        return self.N
    
    def pastPi(self,i):
        """
        Funcion que regresa la posicion pasada de la particula i
        """
        
        return self.pastPositionX[i],self.pastPositionY[i]
    
    def currentPi(self,i):
        """
        Funcion que regresa la posicion actual de la particula i
        """
        
        return self.Particulas[i].positionX(),self.Particulas[i].positionY()
    
    def returnT(self):
        """
        Funcion que regresa la temperatura del sistema
        """
        
        return self.T
```

El sistema anterior representa un sistema con movimiento uniformes, sin embargo la implementacion propuesta no sigue este patron de cambio de estado, si no uno dado por el hamiltoniano de Maxwell-Boltzman. Implementemos ahora el sistema con dicho hamiltoniano siguiendo el esquema propuesto para usar el algoritmo de Metropolis-Hastings

```python
def SistemaSimulacion(L,T,N,n,alpha):
    
    """
    Funcion para simular de un sistema termodinamico
    L =: Longitud del sistema
    T =: Temperatura del sistema
    N =: Numero de particulas
    n =: Numero de simulaciones
    alpha =: Libertad de movimiento
    
    """
    # Definimos la constante de Boltzman
    # Para efectos del proyecto supondremos unidades donde la constante es 1
    kB=1
    
    # Beta termodinamica
    beta=1/(kB*T)
    
    #Creamos el sistema
    SistemaS=Sistema(L,T,N)
    # Inicializamos el sistema
    SistemaS.initSistema()
    
    # Auxiliares
    data=[]
    Energy=[0]
    
    # Corremos la simulacion
    for i in range(n+2):
        # Auxiliares
        perturbation=[]
        it=[0]
        
        # Posicion pasada del sistema
        posX,posY=SistemaS.positions()
        U=stats.uniform(0,1).rvs()
        
        # Iteracion
        for j in range(0,i):
            it.append(j)
            
        # Datos del Sistema    
        data.append([posX,posY,Energy,it])
        
        # Variables de perturbacion
        for k in range(0,SistemaS.returnN()):
            perturbation.append([stats.uniform(-1,2).rvs(),stats.uniform(-1,2).rvs()])
        
        # Perturbamos el sistema
        SistemaS.perturbateParticles(alpha,perturbation)
    
        aux=SistemaS.Energia()
        
        # Paso de aceptacion 
        if(U <= min(1,np.e**(-beta*(aux-Energy[i])))):
            # Aceptamos el cambio de estado
            Energy.append(aux)
            
        else:
            # No aceptamos el cambio de estado
            SistemaS.returnPast()
            # Por lo tanto la energia es 0
            Energy.append(0)
            
    # Regresamos los datos
    return data
```

```python
# Numero de particulas
N=20
# Numero de simulaciones
n=300
# Libertad de movimiento
alpha=0.5
# Temperatura
T=10
# Longitud del sistema
L=10

# Sistema E
E=SistemaSimulacion(L,T,N,n,alpha)
```

<center><h4>VIII. Interpretacion de resultados  </h4></center>

<div style="text-align: justify">Los resultados son claros y corresponden a la intuicion fisica del calor especifico. <br>
Para el caso de un temperatura variable, notamos que las primeras temperaturas tiene un calor especifico bajo, esto es debido a que la temperatura es baja y por lo tanto, segun la ecuacion del calor [6], la difusion del calor es sencilla. Conforme llegamos a una cierta temperatura el coeficiente baja, debido a que la energia del sistema no es proporcional a la temperatura: nos acercamos a la maxima energia termica; y como el calor especifico es inversamente propocial a la temperatua entonces esta tiende a disminuir. Ademas se aprecia que el aumento en la libertad de movimento influye en el calor especifico, esto se atribuye a que un mayor movimiento en las particulas del sistema implica mayor varianza en la energia del sistema. Notemos que la cota para el error disminuye conforme la temperatura aumenta: esto es un efecto de la transformacion que sufre el estimador.<br>
La longitud es similar, un mayor espacio de estados implica que hay mas estados posibles y por lo tanto mayor varianza energetica  y energia requerida para el aumento de temperatura.<br>
Hablando de la libertad de movimiento, el calor especifico sufre un cambio drastico pasado un cierto valor y justamente se relacion con aproximar el maximo nivel energetico del sistema, similar lo que sucede con la temperatura variable.<br>
Por ultimo el numero de particulas, la relacion inversa que tiene el numero de particulas con la estimacion de calor especifico es lo que causa que un numero mayor de particulas impliquen un menor calor especifico.</div>

<div style="text-align: justify">El calor especifico de un sistema es una cantidad de sustancial importancia en la industria; calcular un valor para el calor especifico no es tarea sencilla dada la formalidad de la fisica estadistica, puesto que si quiera obtener un valor para la esperanza energetica del sistema implica integracion multiple. Gracias a los metodos monte-carlo dar estimacion para este valor resulta sencillo, dado que no hay que evaluar integrales multiples, y viable, puesto que la relacion error numerico y costo computacional no es elevada.<br>
